```python
import requests
import bs4
```

# Задача 1
Поэкспериментируйте с парсингом любой страницы на ваш вкус, бегло просмотрите документацию библиотек beautiful soup и requests. При парсинге вам в любом случае потребуется смотреть на html-код страницы, чтобы понять, какие элементы вас интересуют.


```python
req = requests.get('https://scikit-learn.org/stable/modules/classes.html')
parser = bs4.BeautifulSoup(req.text, 'lxml')
result = list(map(lambda x: x.get_text(), parser.findAll('span', attrs={'class':'pre'})))

print("Результат парсинга:")
result
```

# Задача 2
Чтобы продемонстрировать, что вы разобрались, как использовать requests и beautiful soup, распарсите: а) из статьи https://en.wikipedia.org/wiki/Bias-variance_tradeoff все заголовки верхнего уровня; б) со страницы https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms названия всех статей в категории Machine Learning Algorithms

## Пункт 2.1
Распарсите из статьи https://en.wikipedia.org/wiki/Bias-variance_tradeoff все заголовки верхнего уровня;


```python
req = requests.get('https://en.wikipedia.org/wiki/Bias-variance_tradeoff')
parser = bs4.BeautifulSoup(req.text, 'lxml')
result = list(map(lambda x: x.get_text(), parser.findAll('span', attrs={'class':'mw-headline'})))

print("Результат парсинга:")
result
```

## Пункт 2.2
Распарсите со страницы https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms названия всех статей в категории Machine Learning Algorithms

```python
req = requests.get('https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms')
parser = bs4.BeautifulSoup(req.text, 'lxml')
parent = parser.findAll("div", attrs={'id': 'mw-pages'})
result = [x.text for x in parser.select("div#mw-pages div.mw-category-group li a")]

print("Результат парсинга:")
result
```
